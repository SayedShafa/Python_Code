```python
# Menu dictionary: stores food items and their prices.
menu={

"1.Chicken Cheese BUrger":249,
"2.Beef Cheese Burger" :249,
"3.Smoke Chiken CheesebSandwich" :275,
"4.Chicken Sub ":225,
"5.Butter Chicken w/Lemon Smoked Rice":299,
"6.Tandoori Drumstick w/Tandori Rice" :299,
"7.Meat Madness":300,
"8.Chicken Nuggets":300,
"9.Loaded Sausage" :250,
"10.Handcrafted Wedges ":175,
"11.Handcrafted Fries":175,
"12.Masala Potatos":149,
"13.Naga Wings":149,
"14.Choco Coffee":175,
"15.Freege Mango":139,
"16.Cream Soda":139,
"17.Orange Juice":139,
"18.Lassi" :139,
"19.Water 500ml":30,
"20.Water 1L":40,
}

# Display the menu items with prices
for item,price in menu.items():
    print(f"{item}: {price} BDT")

# List to store ordered items
a=[]

# Variable to store total bill amount
bill=0

# Convert dictionary keys and values into lists for index-based access
menukey=list(menu.keys())  
menuvalue=list(menu.values())

try:
    
    # Take first order input from user
    choose=int(input("Choose Your Food(1 to 20) :"))

    # user choice
    if 1<=choose<=20:
    
        # Get selected item name and price by number
        key=menukey[choose-1]
        value=menuvalue[choose-1]
    
    
        a.append(f"({key}:{value} BDT)")    #add item to the list

        print(f"You have oredered {a}")  #show current order
    
        bill+=value  # Add price to total bill

    
    # Handle invalid input 
    else:
       print("Invalid choice")   
        
except ValueError:
    print("Invalid Input!!!!Please enter a number (1 to 20) ")
    

# Ask user if they want to order more
typ=str(input("Want to select more(Type:yes/no)?:"))

# Loop runs until user says 'no'
while typ!="no":

    # If input is not 'yes', exit loop
    if typ!="yes":
        print("Invalid syntex.Try again")
        break

    # Take another order
    else:
        
        try:
            
            choose=int(input("Choose Your Food(1 to 20) :"))

            if 1<=choose<=20:
                key=menukey[choose-1]
                value=menuvalue[choose-1]
             
                a.append(f"({key}:{value} bdt)") # Add new item to order list
                print(f"You Have oredered {a}")   #show updated order
                bill+=value   # Update total bill

         
            else:
                print("Invalid choice")   # Handle invalid input
                
        except ValueError:
            print("Invalid Input!!!!Please enter a number (1 to 20) ")
            
               
    typ=str(input("Want to select more(Type:yes/no)?:")) #asking each time after adding food
    
# Final bill calculation when user finishes ordering      
if typ=="no":

    
    print("\n-------- BILL SUMMARY --------")
    print(f"\nYour Order is {a} .\nYour ordered bill is : {bill} BDT.\nvat 10%.\nYour Total bill : {bill + bill*0.1} BDT.") 
    print("\n------------------------------")

   
    #confirmation
    
    confirm=str(input("\nAre you sure about your order?:"))
    
    if confirm=="yes" and bill!=0:
         print("\nYour order is confirmed! Thank You.")
    elif confirm=="no" :
        print("Your order is not confirmed!Kindly Select Your Food Again.")
    else:
        if bill==0:
            print("You didn't order anything!Kindly Select Your Food Again.")
        else:
            print("Invalid syntex.Try again!")
```
