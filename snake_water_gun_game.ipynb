```python
#snake_water_gun_game

import random


# Define the possible options for the game
options=["snake","gun","water"]

# Take input from the user
a=(input("Take your input from gun , water ,snake:"))

# Randomly select one option for the computer
b=random.choice(options)

print(f"Computer  input:{b}")

# Check for draw condition
if (a=="snake" and b=="snake") or (a=="water" and b=="water") or (a=="gun" and b=="gun") :
    print("Draw")
    
# Check if the user wins
elif (a =="gun" and  b=="snake") or (a=="snake" and b=="water") or (a=="water" and b=="gun" ) :
    print("You wins!")

# Check if the computer wins
elif (b =="gun" and  a=="snake") or (b=="snake" and a=="water") or (b=="water" and a=="gun" ) :
    print("Computer wins!")

#invalid input
else:
    print("Typing wrong")
```
